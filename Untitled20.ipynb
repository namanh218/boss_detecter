```python
images = []
labels = []

def prepare_image(file):
    img_path = ''
    img = image.load_img(img_path + file, target_size=(IMAGE_SIZE, IMAGE_SIZE))
    img_array = image.img_to_array(img) 
    img_array /= 255
#     img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return img_array
# Generate dataset

for path, subdirs, files in os.walk(IMAGE_DATA):
    for name in files:
        if name == '.DS_Store':
            pass
        else:
            img_path = os.path.join(path, name)
#             print(img_path)
            images.append(prepare_image(img_path))
            labels.append(path.split('/')[-1])
#             print(path.split('/')[-1])
```

```python
model = load_model('./model/facenet_keras.h5')
```

```python
model.summary()
```

```python
images = np.array(images)
images.shape
```

```python
y = model.predict(images)
```

```python
labels = list(map(lambda x: 1 if x == 'Minh Anh' else 0, labels))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(y, labels, test_size=0.1)
```

```python
svm_2 = svm.SVC(probability = True)
svm_2.fit(X_train, y_train)
```
